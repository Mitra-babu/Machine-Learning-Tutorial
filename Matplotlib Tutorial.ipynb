```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

# Displaying Dataser

```python
plt.scatter(x,y,c='g')
```

Labeling

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,c='g')
```

# Creating subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g--.')

plt.subplot(2,2,3)
plt.plot(x,y,'b--o')

plt.subplot(2,2,4)
plt.plot(x,y,'cyan')
```

Sign curb

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.title('Sign wave')
```

# Bar Chart

```python
x1=[2,8,10]
y1=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]

plt.bar(x1,y2,color='cyan')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

# Histogram

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

# Box plot

```python
data=[np.random.normal(0,std,100) for std in range(0,5)]
data
```

```python
plt.boxplot(data,vert=True, patch_artist=True)
```

```python
plt.boxplot(data,vert=False, patch_artist=True)
```

# Pie chart

```python
labels='Python','C++','MySQL','Java'
sizes=[300,420,250,290]
colors=['cyan','olive','magenta','lightcoral']
explode=(0.4,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
labels='Python','C++','MySQL','Java'
sizes=[300,420,250,290]
colors=['cyan','olive','magenta','lightcoral']
explode=(0.3,0.1,0.1,0.1)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.5f%%',shadow=False)
plt.show()
```
